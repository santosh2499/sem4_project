```python
data['Category'].replace('Grooming', 'Personal Care',inplace=True)
data['Category'].replace('Beauty', 'Personal Care',inplace=True)
data['Category'].replace('Investment', 'Financial Services',inplace=True)
data['Category'].replace('Apparel', 'Household',inplace=True)
data['Category'].replace('Culture', 'Travel and Culture',inplace=True)
data['Category'].replace('Tourism', 'Travel and Culture',inplace=True)
```

```python
encoder = LabelEncoder()
data['target'] = encoder.fit_transform(data['Category'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['Note'], data['target'], test_size=0.15, random_state=42)
```

```python
X_train_tensor = tf.convert_to_tensor(X_train.values.astype(str), dtype=tf.string)
y_train_tensor = tf.convert_to_tensor(y_train.values, dtype=tf.int32)
X_test_tensor = tf.convert_to_tensor(X_test.values.astype(str), dtype=tf.string)
y_test_tensor = tf.convert_to_tensor(y_test.values, dtype=tf.int32)
```

```python
# # preprocessor = hub.KerasLayer("https://kaggle.com/models/tensorflow/bert/TensorFlow2/en-uncased-preprocess/3")
# # encoder = hub.KerasLayer("https://www.kaggle.com/models/tensorflow/bert/TensorFlow2/bert-en-uncased-l-12-h-768-a-12/2",trainable=True)
model_path = '/content/bert_model'  # Replace with your actual path

# Load the SavedModel
loaded_model = tf.saved_model.load(model_path)

# Create a custom Keras layer
class BertEncoder(tf.keras.layers.Layer):
    def __init__(self, model, **kwargs):
        super(BertEncoder, self).__init__(**kwargs)
        self.model = model

    def call(self, inputs):
        outputs = self.model.signatures['serving_default'](sentences=inputs)
        return outputs['bert_pack_inputs']

# Create the Keras layer
bert_layer = BertEncoder(loaded_model)

# Model definition
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name="text")
outputs = bert_layer(text_input)
outputs_float = tf.keras.layers.Lambda(lambda x: tf.cast(x, tf.float32))(outputs) #wrap tf.cast
l = tf.keras.layers.Dropout(0.1, name="dropout")(outputs_float)
l = tf.keras.layers.Dense(16, activation='softmax', name="output")(l)
model = tf.keras.Model(inputs=[text_input], outputs=[l])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train_tensor, y_train_tensor, epochs=10)
```

```python
model.evaluate(X_test_tensor,y_test_tensor)
```
